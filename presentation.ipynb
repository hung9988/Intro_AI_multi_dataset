```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

# Displaying roc_plot images
fig = plt.figure(figsize=(20, 20))
for i, j in enumerate([0,1,2,3,4]):
    img = mpimg.imread(f'images/roc_plot{j}.png')
    ax = fig.add_subplot(3, 2, i+1)  # 2 rows, 2 columns
    ax.imshow(img)
    ax.set_title(f'dataset{j}')  # Set title for each image
    plt.axis('off')
plt.subplots_adjust(wspace=0, hspace=0.05)  # Remove space between images
plt.show()
```
